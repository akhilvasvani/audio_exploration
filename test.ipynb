```python
import numpy as np
import librosa
import librosa.display
from audiomentations import Compose, AddGaussianNoise, Shift, TimeStretch, PitchShift
from matplotlib import pyplot as plt
```

```python
def plot_transformation(y, sr, transformation_name):
    D = librosa.stft(y)  # STFT of y
    S_db = librosa.amplitude_to_db(np.abs(D), ref=np.max)
    fig, ax = plt.subplots()
    img = librosa.display.specshow(S_db, x_axis='time', y_axis='linear', ax=ax)
    ax.set(title=transformation_name)
    fig.colorbar(img, ax=ax, format="%+2.f dB")

    return plt.gcf()
```

```python
def plot_audio_transformations(y, sr, pipeline: Compose):

    _ = plot_transformation(y, sr, "Original")

    figs = []
    for individual_transformation in pipeline.transforms:
        transformation_name = str(type(individual_transformation)).split("'")[1].split(".")[-1]
        figs.append(plot_transformation(individual_transformation(y, sr), sr, transformation_name=transformation_name))
    return figs
```

```python
def load_audio_sample(file_path: str):
    y, sr = librosa.load(file_path, sr=22050)

    return y, sr, file_path
```

```python
transform = Compose([AddGaussianNoise(), Shift(min_fraction=-0.5, max_fraction=0.5, p=0.5),
                     TimeStretch(min_rate=0.8, max_rate=1.25, p=0.5),
                     PitchShift(min_semitones=-4, max_semitones=4, p=0.5), ])
```

```python
y, sr, _ = load_audio_sample(
    file_path="/Users/pascal/Documents/DeepLearning/datasets/audio/ESC-50-master/audio/1-18810-A-49.wav")
```

```python
figs = plot_audio_transformations(y, sr, transform)
```
